## Get Latlong for each HQ location

```python
import requests
from config import gkey
print(gkey)
```

```python
API_url = "https://maps.googleapis.com/maps/api/geocode/json"
```

```python
tables['lat'] = 0
tables['long'] = 0
```

```python
for index, row in tables.iterrows():
    target = row['Security']+' '+row['Headquarters Location']
    params = {"address": target, "key": gkey}
    response = requests.get(API_url, params=params)
    lat = response.json()["results"][0]["geometry"]["location"]["lat"]
    long = response.json()["results"][0]["geometry"]["location"]["lng"]
    tables.loc[index, 'lat'] = lat
    tables.loc[index, 'long'] = long
    print(index, target, lat, long)
```
